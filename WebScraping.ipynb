```python
def scrape_page(url, tag_name):
    # Peticion GET
    response = requests.get(url)

    # Conexion Exitosa
    if response.status_code == 200:
        # Codigo HTML
        html_content = response.text
        soup = BeautifulSoup(html_content, "html.parser")
        elements = soup.find_all(tag_name)
        elements_list = [element.text for element in elements]
        return elements_list

    else:
        # Conexion Erronea
        print("Error en la petición:", response.status_code)
```

```python
elements = scrape_page("https://www.datosmundial.com/america/mexico/crecimiento-poblacional.php", "td")

for element in elements:
    print(element)
```

 Conclusiones:
Se utiliza una libreria aparte de parcel llamada BeautifulSoup que sirve para extraer y limpian los datos con lo cual pudimos verificar que la captura de la informacion de la pagina fue exitos
